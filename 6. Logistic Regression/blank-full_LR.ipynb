```python
data[['job','marital','education','default','housing','loan','contact','poutcome','month','y']] = data[['job','marital','education','default','housing','loan','contact','poutcome','month', 'y']].astype('category')
```

```python
data.info()
```

```python
#Label Encoding

data['month'] = data['month'].cat.codes
data['job'] = data['job'].cat.codes
data['marital'] = data['marital'].cat.codes
data['education'] = data['education'].cat.codes
data['default'] = data['default'].cat.codes
data['housing'] = data['housing'].cat.codes
data['loan'] = data['loan'].cat.codes
data['contact'] = data['contact'].cat.codes
data['poutcome'] = data['poutcome'].cat.codes
data['y'] = data['y'].cat.codes
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,16]

x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
```

```python
print(y_train.shape)
```

```python
print(x_test.shape)
```

```python
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```
